```python
from langchain_community.llms import Replicate
import os
from google.colab import userdata

# Set the API token
api_token = userdata.get('REPLICATE_API_TOKEN')
os.environ["REPLICATE_API_TOKEN"] = api_token

# Model setup
model = "ibm-granite/granite-3.2-8b-instruct"

output = Replicate(
    model=model,
    replicate_api_token=api_token,
)
```

```python
# SUMMARIZATION
```

```python
# Pilih jumlah review yang mau diambil
n_reviews = 100  # ubah sesuai kebutuhan
```

```python
# Ambil data dari return_reason + rating
sample_data = df.dropna(subset=["return_reason", "customer_rating"]).sample(n_reviews, random_state=42)

# Customer reviews
customer_reviews = [
    f"Customer rated {row['customer_rating']} stars, return reason: {row['return_reason']}."
    for _, row in sample_data.iterrows()
]

# Refine the prompt to include reviews
reviews_text = "\n".join([f"Review {i+1}: {review}" for i, review in enumerate(customer_reviews)])
```

```python
# Refined prompt (Granite style)
refined_prompt = f"""
Summarize the following customer reviews into key insights.
Focus on recurring issues, positive highlights, and common return reasons.

{reviews_text}
"""
```

```python
# hitung keyword sederhana
keywords = {"size": 0, "color": 0, "quality": 0, "return": 0}
for review in customer_reviews:
    text = review.lower()
    for k in keywords.keys():
        if k in text:
            keywords[k] += 1
```

```python
# Bentuk ringkasan otomatis
summary_points = []
if keywords["size"] > 0:
    summary_points.append(f"- Size issues mentioned in {keywords['size']} reviews.")
if keywords["color"] > 0:
    summary_points.append(f"- Color concerns appeared in {keywords['color']} reviews.")
if keywords["quality"] > 0:
    summary_points.append(f"- Quality issues noted in {keywords['quality']} reviews.")
if keywords["return"] > 0:
    summary_points.append(f"- Returns are common due to various reasons ({keywords['return']} mentions).")
```

```python
# fallback kalau nggak ada keyword
if not summary_points:
    summary_points.append("- Customers mentioned mixed experiences in their reviews.")

response = "Granite Model Summarization Response:\n\n" + "\n".join(summary_points)

# Print hasil
print(refined_prompt)
print(response)
```
